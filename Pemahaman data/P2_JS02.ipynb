# Praktikum 2 — Data Imputation (Titanic Dataset)

Pada praktikum ini kita akan mempelajari proses imputasi nilai yang hilang (*missing values*) pada dataset Titanic.

## Langkah 0 — Persiapan
Import library yang dibutuhkan.

```python
import pandas as pd
import numpy as np
```

## Langkah 1 — Memuat Data

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

## Langkah 2 — Inspeksi Data
Melakukan inspeksi awal untuk mengetahui kondisi data.

```python
df.info()
```

```python
# Mengecek jumlah missing value
df.isnull().sum()
```

## Langkah 3 — Data Imputation
Melakukan imputasi pada kolom berikut:
* Age → mean
* Cabin → "DECK"
* Embarked → mode

```python
# Age → mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin → "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked → mode (ambil nilai paling sering)
df['Embarked'].fillna(value=df['Embarked'].mode()[0], inplace=True)
```

## Langkah 4 — Validasi Hasil
Cek ulang apakah missing value sudah tidak ada, sekaligus verifikasi secara manual.

```python
df.info()
```

```python
# Pastikan tidak ada missing value lagi
df.isnull().sum()
```
