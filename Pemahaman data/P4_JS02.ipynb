```python
import cv2
import matplotlib.pyplot as plt
```

## Langkah 1 — Load Gambar dan Ubah ke RGB
OpenCV membaca gambar dalam format **BGR**, sehingga harus dikonversi menjadi **RGB** agar warna tampil benar.

```python
# Membaca gambar (BGR default di OpenCV)
img = cv2.imread("Lenna.jpg")

# Konversi BGR → RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Tampilkan gambar
plt.imshow(img_rgb)
plt.title("Gambar Asli (RGB)")
plt.axis("off")
plt.show()
```

## Langkah 2 — Resize Gambar (128 × 128 px)
Proses resize merupakan bagian umum dari pra pengolahan data citra sebelum memasukkan citra ke model pembelajaran mesin.

```python
# Resize gambar ke ukuran 128x128
img_resized = cv2.resize(img_rgb, (128, 128))

plt.imshow(img_resized)
plt.title("Gambar setelah Resize (128x128)")
plt.axis("off")
plt.show()
```

## Langkah 3 — Ubah ke Grayscale
Grayscale sering digunakan untuk mengurangi kompleksitas citra dan menghemat komputasi.

```python
# Konversi ke grayscale
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, cmap="gray")
plt.title("Gambar Grayscale")
plt.axis("off")
plt.show()
```

### 📌 Kesimpulan
Pada praktikum ini kita telah melakukan tahapan pra pengolahan citra:
- Load gambar & konversi warna BGR → RGB
- Resize citra menjadi 128 × 128
- Konversi citra menjadi Grayscale

Tahapan selanjutnya dalam pembelajaran mesin dapat mencakup ekstraksi fitur dan pelatihan model.
