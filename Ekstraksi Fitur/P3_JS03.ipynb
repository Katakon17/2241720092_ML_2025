Pada praktikum ini kita melakukan Encoding, Imputasi, Standardisasi, dan Seleksi fitur menggunakan data Titanic.

```python
#### Langkah 0 - Load Library
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#### Langkah 1 - Load Data
dpath = "Titanic-Dataset-selected.csv"
df = pd.read_csv(dpath)
df.head()
```

```python
#### Langkah 2 - Encoding fitur 'Sex'
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
print("=== Hasil Setelah Encoding 'Sex' ===")
print(df[['Sex']].head(), "\n")
```

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['Sex'] = le.fit_transform(df['Sex']) # proses encoding
df['Cabin'] = le.fit_transform(df['Cabin']) # proses encoding
```

```python
#### Langkah 4 - Imputasi Data pada Usia
df['Age'] = df[['Age']].fillna(df['Age'].median())
print("=== Hasil Setelah Imputasi Median pada 'Age' ===")
print(df[['Age']].head(10), "\n")
```

```python
#### Langkah 5 - Standardisasi fitur 'Age'
scaler = StandardScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
print("=== Hasil Setelah Standardisasi 'Age' ===")
print(df[['Age']].head(10), "\n")
```

```python
#### Langkah 6 - Cek hasil akhir preprocessing
print("=== Hasil Preprocessing (5 baris pertama) ===")
df.head()
```

```python
#### Langkah 7 - Seleksi Fitur (5 variabel penting)
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']]
df.head()
```
