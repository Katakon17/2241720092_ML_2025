- Load gambar
- Ubah colorspace BGR → RGB
- Resize gambar
- Konversi ke grayscale

```python
import cv2
import matplotlib.pyplot as plt
```

## Langkah 1 — Load Gambar
Load gambar lalu konversi dari BGR ke RGB agar warna tampil dengan benar.

```python
# Membaca gambar (default OpenCV: BGR)
img = cv2.imread("Lenna.jpg")

# Konversi BGR → RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Gambar Asli (RGB)")
plt.axis("off")
plt.show()
```

## Langkah 2 — Resize Gambar
Ubah ukuran gambar menjadi **128 × 128 px**.

```python
# Resize gambar menjadi 128x128
img_resized = cv2.resize(img_rgb, (128, 128))

plt.imshow(img_resized)
plt.title("Gambar Setelah Resize (128x128)")
plt.axis("off")
plt.show()
```

## Langkah 3 — Konversi ke Grayscale
Konversi citra ke format grayscale.

```python
# Konversi RGB → Grayscale
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap="gray")
plt.title("Gambar Grayscale")
plt.axis("off")
plt.show()
```
