### ðŸ”¹ Langkah 3 â€“ Imputasi Missing Value
Strategi imputasi yang digunakan:
- **Age** â†’ Mean
- **Cabin** â†’ Substitusi string "DECK"
- **Embarked** â†’ Mode

```python
# Age â€“ imputasi mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin â€“ substitusi "DECK"
df['Cabin'].fillna(value='DECK', inplace=True)

# Embarked â€“ mode
df['Embarked'].fillna(value=df['Embarked'].mode()[0], inplace=True)
```

### ðŸ”¹ Langkah 4 â€“ Validasi Hasil Imputasi

```python
df.info()
```

### ðŸ”¹ Langkah 5 â€“ Ekstraksi Fitur
Fitur baru yang ditambahkan:
- **FamilySize** â†’ `SibSp + Parch + 1`
- **Title** â†’ Ekstraksi gelar dari kolom *Name*
- **AgeBin** â†’ Kategori usia
- **CabinDeck** â†’ Huruf pertama kolom *Cabin*
- **FarePerPerson** â†’ `Fare / FamilySize`

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. FamilySize
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# 2. Title
df['Title'] = df['Name'].str.extract(r',\s*([^\.]+)\.', expand=False)
title_mapping = {
    'Mlle': 'Miss','Ms': 'Miss','Mme': 'Mrs',
    'Lady': 'Rare','Countess': 'Rare','Capt': 'Rare','Col': 'Rare',
    'Don': 'Rare','Dr': 'Rare','Major': 'Rare','Rev': 'Rare',
    'Sir': 'Rare','Jonkheer': 'Rare','Dona': 'Rare'
}
df['Title'] = df['Title'].replace(title_mapping)

# 3. AgeBin
df['AgeBin'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior'])

# 4. CabinDeck
df['CabinDeck'] = df['Cabin'].astype(str).str[0]

# 5. FarePerPerson
df['FarePerPerson'] = df['Fare'] / df['FamilySize']

df[['Name','Title','Age','AgeBin','FamilySize','Cabin','CabinDeck','FarePerPerson']].head()
```

### ðŸ”¹ Langkah 6 â€“ Visualisasi Fitur Hasil Ekstraksi

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate berdasarkan Family Size')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='Title', y='Survived', data=df, order=df.groupby('Title')['Survived'].mean().sort_values().index)
plt.title('Survival Rate berdasarkan Title')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='AgeBin', y='Survived', data=df, order=['Child','Teen','YoungAdult','Adult','Senior'])
plt.title('Survival Rate berdasarkan Kelompok Usia (AgeBin)')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='CabinDeck', y='Survived', data=df, order=df.groupby('CabinDeck')['Survived'].mean().sort_values().index)
plt.title('Survival Rate berdasarkan Cabin Deck')
plt.show()
```
